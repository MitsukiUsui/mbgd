```python
import pandas as pd
```

```python
strainFilepath="../preprocess/strain.lst"
strain_lst=[s.strip() for s in open(strainFilepath, 'r').readlines()]
familyFilepath="family.lst"
family_lst=[s.strip() for s in open(familyFilepath, 'r').readlines()]
```

```python
columns_lst=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", 
                             "qstart", "qend", "sstart","send", "evalue", "bitscore"]
dct_lst=[]
    
for strain in strain_lst:
    for family in family_lst:
        filepath="../blastn/result/{0}_{1}.tab".format(strain,family)
        if len(open(filepath, 'r').readlines())>0:
            df=pd.read_csv(filepath,delimiter='\t', header=None)
            for _,row in df.iterrows():
                dct={}
                dct["strain"]=strain
                dct["family"]=family
                
                for i,column in enumerate(columns_lst):
                    dct[column]=row[i]
                
                dct_lst.append(dct)
```

```python
out_df=pd.DataFrame(dct_lst)
out_df=out_df[["strain","family"]+columns_lst]
out_df.index.name="region_id"
print(out_df.shape)
out_df.head(10)
```

```python
out_df.to_csv("./out/summarize.csv")
```
